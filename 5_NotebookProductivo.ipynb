```python
import pickle
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
filaname="ModeloFinalCredit.pkl"
```

```python
with open(filaname, 'rb') as file:
    Modelo = pickle.load(file)
```

```python
Modelo
```

```python
clientes=pd.read_csv("BaseNueva.csv")
```

```python
clientes
```

```python
Predicciones=Modelo.predict(clientes)
Probabilidades=Modelo.predict_proba(clientes)
```

```python
clientes["score"]=Probabilidades[::,1]
clientes["Prediccion"]=Predicciones
```

```python
clientes.head()
```

```python
data_sorted = clientes.sort_values(by='score', ascending=True)
data_sorted
```

```python
# Calcular el número de usuarios por decil
num_users_per_decile = len(data_sorted) // 10
# Crear una nueva columna 'decil' para indicar el decil al que pertenece cada usuario
data_sorted['decil'] = (data_sorted.reset_index().index // num_users_per_decile) + 1
```
